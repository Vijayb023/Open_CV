```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
```

```python
Image = mpimg.imread("0.jpg")
imgplot = plt.imshow(Image)
brow = mpimg.imread("brow.png")


```

```python
brow = cv2.resize(brow,(300,100))
print("image Dimension ={}".format(brow.shape))
```

```python
browBGR = brow[:,:,0:3]
browMask1 = brow[:,:,3]

```

```python
plt.figure(figsize=[15,15])
plt.subplot(121);plt.imshow(browBGR[:,:,::-1]);plt.title('Brow Color channels');
plt.subplot(122);plt.imshow(browMask1,cmap='gray');plt.title('Brow Alpha channel');
```
